```python
import numpy as np
```

```python
import matplotlib.pylab as plt
```

```python
# Establecer las condiciones básicas
bmax = 1
# Normalizar bmax a 1
freq = 60 # 60 Hz
w = 2*np.pi*freq #  velocidad angular (rad/s)
```

```python
# Primero, generar los tres campos magnéticos

t = np.linspace(0,1/60,6000)

Baa_R = np.sin(w*t) * np.cos(0)
Baa_I = np.sin(w*t) * np.sin(0)
Bbb_R = np.sin(w*t + 2*np.pi/3) * np.cos(2*np.pi/3) 
Bbb_I = np.sin(w*t + 2*np.pi/3) * np.sin(2*np.pi/3)
Bcc_R = np.sin(w*t + 2*np.pi/3) * np.cos(-2*np.pi/3)
Bcc_I = np.sin(w*t + 2*np.pi/3) * np.sin(-2*np.pi/3)
```

```python
# Calcular Bnet
Bnet_R = Baa_R + Bbb_R + Bcc_R
Bnet_I = Baa_I + Bbb_I + Bcc_I

# Calcular un círculo que represente el valor máximo esperado de Bnet
circle_R = 1.5 * np.cos(w*t) 
circle_I = 1.5 * np.sin(w*t)
```

```python
plt.figure(figsize=(7,6))
plt.plot(Bnet_R, Bnet_I, label="$B_{net}$")
plt.plot(Baa_R, Baa_I, label="$B_{aa}$")
plt.plot(Bbb_R, Bbb_I, label="$B_{bb}$")
plt.plot(Bcc_R, Bcc_I, label="$B_{cc}$")
plt.xlabel("Real", fontsize=15)
plt.ylabel("Imaginary", fontsize=15)
plt.legend(fontsize=15)
plt.plot(circle_R, circle_I, '--k')
```
